```python
top_cities= df["City"].value_counts().head(10)
plt.figure(figsize=(10,5))
top_cities.plot(kind="bar")
plt.title("Top 10 Cities")
plt.xlabel("City")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
