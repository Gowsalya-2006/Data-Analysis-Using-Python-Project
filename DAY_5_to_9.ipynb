DAY-4 TASK:
      PRODUCT SALES PERFORMENCE ANALYS

```python
import numpy as np
products = ["Laptop","Mobile","Tablet","Headphone"]
sales = np.array([120,250,50,150])
print("Total Sales",np.sum(sales))

print("Average Sales",np.mean(sales))

print("Highest Sales",np.max(sales))

print("Lowest Sales",np.min(sales))

best_index = np.argmax(sales)
print("Best Selling Product",products[best_index])

print("\nProducts Sales Report")
for i in range(len(products)):
     print(f"{products[i]}:{sales[i]}")
```

Employee Dataset Exploration: DAY_5 TASK

```python
import pandas as pd

employees = {
    "Emp_ID":[101,102,103,104,105],
    "Name":["Arun","Priya","Aarthi","Agalya","Divya"],
    "Department":["IT","HR","FINANCE","IT","HR"],
    "Salary":[45000,50000,55000,60000,65000],
    "Exprience":[2,3,4,5,6]
}
df = pd.DataFrame(employees)
print("Employees Dataset:")
print(df)

print("\nAverage Salary:")
print(df["Salary"].mean())

print("\nHightest Salary:")
print(df["Salary"].max())

print("\nEmployees in HR:")
print(df[df["Department"] == "HR"])

print("\nEmployeesmwith Exprience > 3:")
print(df[df["Exprience"] > 3])

print("\nEmployee Count by Department:")
print(df["Department"].value_counts())

```

```python
import pandas as pd
data = {
    "Region":["South","North","East","west","South","North"],
    "Product":["Laptop","Mobile","Laptop","Tablet","Mobile","Tablet"],
    "Sales":[50000,30000,45000,20000,35000,40000]
}
df = pd.DataFrame(data)
print(df)
```

```python
south_sales = df[df["Region"]=="South"]
print(south_sales)
```

```python
high_sales = df[df["Sales"] > 30000]

print(high_sales)
```

```python
sorted_df = df.sort_values("Sales")

print(sorted_df)
```

```python
region_sales = df.groupby("Region")["Sales"].sum()

print(region_sales)
```

```python
product_sales = df.groupby("Product")["Sales"].sum()

print(product_sales)
```

DAY-7 *TASK*:REGION-WISE SALES REPORT

```python
import pandas as pd
data ={
    "Region":["North","South","North","South","East","West"],
    "Product":["Laptop","Mouse","Keyboard","Laptop","Monitor","Laptop"],
    "Sales":[10000,20000,30000,40000,50000,60000]
}
df = pd.DataFrame(data)
df_active=df[df["Sales"]>0]

df_sorted=df_active.sort_values(["Region","Sales"],ascending=[True,False])

report=df_sorted.groupby("Region")["Sales"].agg(["sum","mean","max","count"])

print(report)

```

MONTHLY SALES DASHBOARD

```python
import matplotlib.pyplot as plt
months=["Jan","Feb","Mar","Apr","May","June"]
sales=[10000,12000,15000,13000,18000,20000]

plt.figure(figsize=(18,5)) # Adjust figure size to accommodate three plots

# Line chart
plt.subplot(1,3,1) # 1 row, 3 columns, first plot
plt.plot(months,sales,marker="o")
plt.title("Sales Trend")
plt.xlabel("Months")
plt.ylabel("Sales")

# Bar chart
plt.subplot(1,3,2) # 1 row, 3 columns, second plot
plt.bar(months,sales,color="red")
plt.title("Sales Comparison")
plt.xlabel("Months")
plt.ylabel("Sales")

# Pie chart
expenses = [4000,7000,1000]
categories = ["Food","Travel","Shopping"]
plt.subplot(1,3,3) # 1 row, 3 columns, third plot
plt.pie(expenses,
        labels=categories, autopct="%1.1f%%")
plt.title("Monthly Expenses")

plt.tight_layout() # Adjust layout to prevent overlapping titles/labels
plt.show()
```
