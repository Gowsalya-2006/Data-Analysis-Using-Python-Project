If and Elif

```python
mark_1=int(input("Enter mark 1:"))
mark_2=int(input("Enter mark 2:"))
mark_3=int(input("Enter mark 3:"))
Total = mark_1+mark_2+mark_3
Average=Total/3
print("Total:",Total)
print("Average:",Average)
if(Average>=50):
  print("pass")
else:
  print("fail")
```

*TASK*-1 Grading system


```python
marks=int(input("Enter mark:"))
if marks >= 90:
     print("Grade:A")
elif marks >= 75:
     print("Grade:B")
elif marks >= 50:
     print("Grade:C")
else:
     print("fail")
```
